1. Create a simple page with some div tags and show different ways to add CSS as well as what happens
when you target the same elements with inline, internal, and external CSS. Also, utilize comments in the
project where required.


Index.html

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Types of CSS</title>

    <link rel="stylesheet" href="Style1.css">


    <style>

        /* Internal CSS  */

        .box {

            color: rgb(2, 10, 24);
            background-color: rgb(155, 209, 211);
            padding: 10px;
            text-align: center;
            line-height: 20px;
            font-size: large;
             border: 2px dotted rgb(2, 10, 24);
             padding: 10px;
             margin: 20px;

        }

 </style>

</head>
<body>

    <!-- This div uses all 3 types of CSS -->
    <div class="box" style="background-color: rgb(221, 94, 164); border: dashed;">

             This is div tag line 1
            <!--Using Inline CSS , Inline css has highest priority that is why it overrides internal & external CSS-->

    </div>


    <!--This div usses Internal & External CSS-->
    <div class="box">

        This is div tag line 2
        <!--Internal CSS overrides External Css as internal css has higher priority than external. -->

    </div>


    <!--This div uses only external CSS-->
    <div class="external-box">
        This is div tag line 3
    </div>


</body>
</html>
```

style1.css

```python
/* External CSS */


/* for box class  */
.box {

    background-color: rgba(137, 43, 226, 0.719);
    color: black;
    border: 3px dotted black;

}


/* for external-box class */
.external-box {

        padding: 10px;
        background-color: lightgreen;
        color: rgb(24, 85, 16);
        margin: 20px;
        border: solid;
}
```

2. Build an HTML page with multiple paragraphs, each assigned a unique class name. Write CSS rules using class selectors to apply distinct styling to each paragraph. Follow the BEM naming convention and explain how you've named the classes.

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>BEM CSS </title>


    <style>
        /* Base block: article */
        .article__paragraph_1--orange {
            font-family: Arial, sans-serif;
            margin: 10px 0;
            padding: 10px;
            border-radius: 50px;
            background-color: rgb(248, 195, 97);
            color: rgb(121, 82, 12);
        }

        /* Modifier: purple background */
        .article__paragraph_2--purple {
            background-color: #db8adb;
            color: rgb(107, 11, 86);
            border-radius: 50px;
            margin: 10px 0;
            padding: 10px;

        }

        /* Modifier: pink background */
        .article__paragraph_3--pink {
            background-color: #e06699;
            color: rgb(167, 15, 66);
             border-radius: 50px;
             margin: 10px 0;
            padding: 10px;
        }

        /* Modifier: blue background */
        .article__paragraph_4--blue {
            background-color: #ddeeff;
            color: #4174a7;
            font-size: large;
             border-radius: 50px;
            margin: 10px 0;
            padding: 10px;
        }


    </style>
</head>

<body>

    <h1 style="text-align: center; color: #004080; ">Paragraph Styling with BEM</h1>

    <!-- BEM: Block = article, Element = paragraph, Modifier = red -->
    <p class="article__paragraph_1 article__paragraph_1--orange">
        This is the first paragraph with a red background.
    </p>

    <!-- BEM: Modifier = purple -->
    <p class="article__paragraph_2 article__paragraph_2--purple ">
        This is the second paragraph with a green background.
    </p>

    <!-- BEM: Modifier = pink-->
    <p class="article__paragraph_3 article__paragraph_3--pink ">
        This is the third paragraph with a blue background.
    </p>

    <!-- BEM: Modifier = blue + large -->
    <p class="article__paragraph_4 article__paragraph_4--blue">
        This is the fourth paragraph with green background and large text.
    </p>

</body>
</html>
```

3.  Develop an HTML form with various input elements. Use CSS to style the form, including setting background colors for input fields. Create a custom color palette for the form elements, and demonstrate how to apply opacity to one of the form sections.

```python
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Styled HTML Form</title>


    <style>

        :root {

        --background-color: rgb(237, 250, 241) ;
        --input-border: rgb(158, 14, 74) ;
        --label-color:#333;
        --accent-color: #72c5ac;
        --primary-color: #ad114d;
        --input-bg: #ffffff ;


         }


    body {

        font-family: Arial, sans-serif;
        background-color: var(--background-color);
        margin: 20px;

    }

    h1 {
      color: var(--primary-color);
      text-align: center;
    }

    form {
      max-width: 600px;
      margin: auto;
      background: #fff;
      padding: 20px;
      border: 2px solid var(--primary-color);
      border-radius: 8px;
      box-shadow: 0 4px 10px rgba(0,0,0,0.1);
    }

    label {
      display: block;
      margin: 10px 0 5px;
      color: var(--label-color);
      font-weight: bold;
    }

    input, textarea, select {
      width: 100%;
      padding: 10px;
      background-color: var(--input-bg);
      border: 1px solid var(--input-border);
      border-radius: 5px;
      margin-bottom: 15px;
      font-size: 1em;
      color: rgba(0, 0, 0, 0.712);
    }

    input[type="radio"],
    input[type="checkbox"] {
      width: auto;
      margin-right: 10px;
    }

    fieldset {
      border: 1px solid var(--accent-color);
      padding: 15px;
      border-radius: 6px;
      margin-bottom: 20px;
    }

    /*  Opacity demonstration */
    .semi-transparent {
      background-color: var(--accent-color);
      opacity: 0.7;
    }

    button {
      background-color: var(--primary-color);
      color: #fff;
      padding: 10px 20px;
      border: none;
      border-radius: 4px;
      cursor: pointer;
    }

    button:hover {
      background-color: #2980b9;
    }

    </style>

</head>
<body>

    <h1> Feedback and Queries</h1>

    <form>

        <!--Section 1 -->
        <fieldset>
            <legend> Personal Information </legend>

           <label for="name"> Name:</label>
           <input type="text" id="name" name="name" placeholder="Your full name">

           <label for="email">Email:</label>
           <input type="email" id="email" name="email" placeholder="Your email address">

           <label for="Password">Password:</label>
           <input type="password" id="password" name="password" placeholder="Create a password">

        </fieldset>


        <!--Section 2  -->
        <fieldset class="semi-transparent">

            <legend>
                Preferences (with opacity)
            </legend>

            <label> Gender:</label>
            <input type="radio" id="male" name="gender" value="male">
             <label for="male">Male</label>

            <input type="radio" id="female" name="gender" value="female">
            <label for="female">Female</label>

            <label for="newsletter">Subscribe to newsletter:</label>
            <input type="checkbox" id="newsletter" name="newsletter">


            <label for="country">Country:</label>
            <select id="country" name="country">
                <option value="">Select...</option>
                <option value="in">India</option>
                <option value="us">United States</option>
                <option value="cn">China</option>
                <option value="rus">Russia</option>

             </select>

        </fieldset>


        <!--Section 3 -->
        <fieldset>
            <legend>Feedback</legend>
            <label>Your Message:</label>
            <textarea id="message" name="message" rows="5" placeholder="Type your message here...."> </textarea>
        </fieldset>

        <button type="submit">Submit</button>


    </form>

</body>
</html>
```
